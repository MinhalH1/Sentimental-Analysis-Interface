```python
import torch
from transformers import AutoModelForSequenceClassification, AutoTokenizer

# Load pre-trained model and tokenizer
model_name = "distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Save the model
model.save_pretrained("./sentiment_model")
tokenizer.save_pretrained("./sentiment_tokenizer")
```

```python
# Step 3 - Create a Gradio Interface
import gradio as gr
import torch
from transformers import AutoModelForSequenceClassification, AutoTokenizer

# Load the saved model and tokenizer
model_name = "./sentiment_model"
tokenizer_name = "./sentiment_tokenizer"
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(tokenizer_name)

# Define the inference function
def predict_sentiment(text):
    inputs = tokenizer(text, return_tensors="pt")
    with torch.no_grad():
        outputs = model(**inputs)
    scores = outputs[0][0].detach().numpy()
    scores = torch.nn.functional.softmax(torch.tensor(scores), dim=0).numpy()
    labels = ["Negative", "Positive"]
    return {labels[i]: float(scores[i]) for i in range(len(labels))}

# Create the Gradio interface
iface = gr.Interface(fn=predict_sentiment, inputs="text", outputs="label")
iface.launch()
```
